```python
import pandas as pd
from matplotlib import pyplot as plt
import plotly.express as px
from datetime import datetime, date, time
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/ My \Drive/Diplomado

d=pd.read_csv('players_in_chess2.csv')
```

```python
from re import A
a=pd.DataFrame(d)
a=a.drop(columns=['Unnamed: 0', 'o_title', 'name', 'fideid', 'flag'])
a['birthday']=2024-a['birthday']

a['category']=np.nan

for i in range(182533):
  if a['birthday'].loc[i]>=50:
    a['category'].loc[i]='Senior'

  if a['birthday'].loc[i]<18:
    a['category'].loc[i]='Sub 18'

  if a['birthday'].loc[i]<50 and a['birthday'].loc[i]>= 18:
    a['category'].loc[i]='Sub 50'
```

```python
b=pd.DataFrame(a.groupby(['sex']).category.count()).reset_index()
b
fig = px.bar(b, x="sex", y="category")
fig.show()
```

```python
c=pd.DataFrame(a.groupby(['sex', 'category']).rating.max()).reset_index()
c
fig = px.bar(c, x="category", y="rating", color="sex", facet_col="sex", facet_col_wrap=2 )
fig.show()
```

```python
d=pd.DataFrame(a.groupby(['country']).rating.max()).reset_index()
d
fig = px.bar(d, x="country", y="rating")
fig.show()
```

```python
e=a.loc[a['country'] == 'MEX']

f=pd.DataFrame(e.groupby(['sex', 'birthday']).rating.value_counts()).reset_index()
fig = px.scatter(f, x="birthday", y="rating", color="sex", facet_col="sex", facet_col_wrap=2 )
fig.show()

g=pd.DataFrame(e.groupby(['sex']).category.value_counts(normalize=True)).reset_index()

fig = px.bar(g, x="sex", y="proportion", color="category", facet_col="category", facet_col_wrap=2 )
fig.show()

g
# import seaborn as sns

# sns.distplot(f)
# plt.hist(f, density=True)
# plt.xlabel("birthday")
# plt.ylabel("rating")
# plt.title("Densidad")
# plt.show()
```
