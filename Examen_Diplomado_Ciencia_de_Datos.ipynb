1. ¿Cuál es la diferencia entre Data Analysis y Data Science?

**R.** La diferencia radica en que *el análisis de datos* busca dar una interpretación a bases de datos y *la ciencia de datos* refiere más a la utilización de éstos para la creación de modelo de predicción o monitoreo

6. ¿Qué es Reducción de Dimensiones y cuáles son sus beneficios?

**R.** Si se tiene un dataset cuyas variables son demasiadas, no será tan fácil trabajar con estas, por lo que con métodos tales como PCA toda la información contenida en una tabla se puede reducir (idealmente) a 2 o 3 variables (no forzosamente).
```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
df_tr=pd.read_csv('/content/train.csv')
df_te=pd.read_csv('/content/test.csv')
df_tr
```

```python
df_te
```

```python
df_tr.drop(['Name', 'Ticket', 'Fare',	'Cabin','Embarked', 'PassengerId'], axis=1, inplace=True)
df_te.drop(['Name', 'Ticket', 'Fare',	'Cabin','Embarked', 'PassengerId'], axis=1, inplace=True)
```

```python
df_tr['Age'].fillna(df_tr['Age'].mean(), inplace=True)
df_te['Age'].fillna(df_te['Age'].mean(), inplace=True)
```

```python
df_tr.info()
df_tr.describe()
```

```python
df_te.info()
df_te.describe()
```

```python
fig=px.bar(df_tr, x='Sex', y='Survived')
fig.show()
```

```python
fig=px.bar(df_tr, x='Parch', y='Survived')
fig.show()
```

```python
df_tr['Sex']=df_tr['Sex'].map({'male':0, 'female':1})
df_te['Sex']=df_te['Sex'].map({'male':0, 'female':1})
```

```python
corr_matrix = df_tr.corr()

fig = px.imshow(corr_matrix, color_continuous_scale = 'RdBu', range_color=[-1,1], text_auto = True, aspect = "auto", labels=dict(color="Correlation"), title='Feature Correlation', height = 800)
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

```python
tar=df_tr['Survived']
df_tr.drop(['Survived'], axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_tr, tar, test_size=0.2, random_state=7)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train_scaled, y_train)
```

```python
rf_pred = rf.predict(X_test_scaled)
```

```python
rf_report = classification_report(y_test, rf_pred)
print("Random Forest Classification Report:")
print(rf_report)
```

```python
def plot_confusion_matrix(cm, labels):
    fig_cm = px.imshow(cm, labels=dict(x="Predicted", y="Actual", color="Count"),
                       x=labels, y=labels, color_continuous_scale='Viridis', text_auto = True,
                       title="Confusion Matrix")
    fig_cm.update_layout(coloraxis_showscale=False)
    fig_cm.show()
```

```python
rf_cm = confusion_matrix(y_test, rf_pred)
plot_confusion_matrix(rf_cm, ['No sobrevive', 'Sobrevive'])
```

```python
X_test_scaled = scaler.transform(df_te)
rf_pred = rf.predict(X_test_scaled)
```

```python
resultados=df_te.copy()
resultados['Survived']=rf_pred
resultados
```

```python
fig=px.bar(resultados, x='Sex', y='Survived')
fig.show()
```
